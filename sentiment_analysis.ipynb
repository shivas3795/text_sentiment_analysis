```python
# syllable count per word
#removing suffix
def strip_suffixes(s,suffixes): 
    for suf in suffixes: 
        if s.endswith(suf): 
            return s.rstrip(suf) 
    return s 

for file in os.listdir(os.getcwd()+'/files'):
    with open(f"{'./files'}\{file}") as g:
            contents = g.read()
            fd = contents.lower().split(' ')
            wr = [x for x in fd]
            total_word = len(wr)
            #print(total_word)
            suf = ['es','ed']
            without_suffix = [strip_suffixes(name,suf) for name in wr]
            #print(without_suffix)
            syllable = [number_of_vowels(ws) for ws in without_suffix]
            #print(sum(syllable))
            #print(syllable)
            row,_ = file.split('.')
            column = 'M'
            save_output(column,int(row)+1,sum(syllable)/total_word)
```

```python
#Percentage of Complex words = the number of complex words /the number of words 
for file in os.listdir(os.getcwd()+'/files'):
    with open(f"{'./files'}\{file}") as g:
            contents = g.read()
            fd = contents.lower().split(' ')
            wr = [x for x in fd]
            total_word = len(wr)
            #print(total_word)
            complex_word = [number_of_vowels(name) for name in wr if number_of_vowels(name)>2]
            row,_ = file.split('.')
            column = 'H'
            save_output(column,int(row)+1,len(complex_word)/total_word)
```

```python
# removing the stop words (using stopwords class of nltk package).
#removing any punctuations like ? ! , . from the word before counting.

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

for file in os.listdir(os.getcwd()+'/files'):
    with open(f"{'./files'}\{file}") as g:
        contents = g.read()
        word_tokens = [token for token in word_tokenize(contents) if token not in stop_words]
        words_count = [tok.lower() for tok in word_tokens if tok.isalpha()]
        len_words_count = len(words_count)
        row,_ = file.split('.')
        column = 'L'
        save_output(column,int(row)+1,len_words_count)
        
```

```python
def Subjectivity(pos,neg,Words_after_cleaning):
    Subjectivity_Score = (pos + neg)/ ((Words_after_cleaning) + 0.000001)
    return Subjectivity_Score
```

```python
#Subjectivity_Score
for cell in range(0,len(df_output)-1): 
    file = str(df['URL_ID'][cell])+'.txt'
    with open(f"{'./files'}\{file}") as g:
        contents = g.read()
        split_words = [word.lower() for word in contents.split()]
        removed_stopwords = [i for i in split_words if i not in stopword]  
        Words_after_cleaning = len(removed_stopwords)

        file_path = './Output Data Structure.xlsx'
        wb = load_workbook(file_path)
        ws = wb['Sheet1']  # or wb.active
        row = 'F'+str(cell+2)
        ws[row] = Subjectivity(df_output['POSITIVE SCORE'][cell],df_output['NEGATIVE SCORE'][cell],Words_after_cleaning)
        wb.save(file_path)
```

```python
#function for calculating sentiment score
def score(data,sentiment):
    score = [w for w in data if w in sentiment]
    return len(score)
```
